```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

# define image size and the model path
IMG_HEIGHT, IMG_WIDTH = 240,670
# Region of Interest (ROI) coordinates
x_start, x_end, y_start, y_end = 600, 1270, 60, 300
model_path = 'E:\\Model\\model_20_20230615-104936.h5'
test_dir = 'E:\\Data\\TestSet'
processed_test_dir = 'E:\\Data\\ProcessedTestSet'

# perform the same preprocessing as prior to training
if not os.path.exists(processed_test_dir):
    os.makedirs(processed_test_dir)

# read all the images in the Testing folder
for i, file in enumerate(os.listdir(test_dir)):
    # read the image
    img = cv2.imread(os.path.join(test_dir, file))

    # Convert the image from BGR to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # crop the image to ROI
    img_cropped = img[y_start:y_end, x_start:x_end]

    # save the image
    cv2.imwrite(os.path.join(processed_test_dir, file), img_cropped)

# load the model
model = load_model(model_path)

# use model to classify the images
for i, file in enumerate(os.listdir(processed_test_dir)):
    # read the image
    img = image.load_img(os.path.join(processed_test_dir, file), target_size=(IMG_HEIGHT, IMG_WIDTH))
    # convert the image to array
    img = image.img_to_array(img)
    # expand the dimensions to match the input dimensions of the model
    img = np.expand_dims(img, axis=0)
    # normalize the image
    img = img/255

    # predict the class
    prediction = model.predict(img)
    # get the class with the highest probability
    class_ = np.argmax(prediction, axis=1)
    if class_ == 0:
        class_ = "Just Saturated"
    elif class_ == 1:
        class_ = "Over Saturated"
    elif class_ == 2:
        class_ = "Under Saturated"
    # get the probability
    prob = np.max(prediction, axis=1)

    # print the class and probability
    print(f"Image: {file} \t Class: {class_} \t Probability: {prob}")

    # plot the image
    plt.imshow(img[0])
    plt.show()
```
